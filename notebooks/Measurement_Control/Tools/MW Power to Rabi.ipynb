```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def determine_ideal_mw_power(v0, p0, v1):
    p1 = p0 - 20*np.log10(v0/v1) 
    return np.round(p1,2)

v1s = np.linspace(1/500e-9, 1/10e-9, 1/1e-4)

y = determine_ideal_mw_power(1/(97e-9*2), 20, v1s)
plt.ylabel('dBm')
plt.xlabel('us')
plt.plot(1/v1s*1e6, y, label='V1')

plt.legend()
plt.show()
```
